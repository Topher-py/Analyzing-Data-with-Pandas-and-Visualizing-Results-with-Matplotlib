```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv('todays_data.csv')
print (data.head())
```

```python
print (data.dtypes)
```

```python
print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
```

```python
print (data.describe())
```

```python
grouped = data.groupby('First Name ')['Test Scores'].mean()
print(grouped)
```

```python
# Bar plot for average test score by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='Test Scores', data=data, palette='Set2')

plt.title("Average Test Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Test Scores")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
# Bar plot for average test score by academic level
plt.figure(figsize=(10, 5))
sns.barplot(x='Academic Level', y='Test Scores', data=data, palette='Set3')

plt.title("Average Test Score by Academic Level")
plt.xlabel("Academic Level")
plt.ylabel("Average Test Score")
plt.ylim(0, 100)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['Test Scores'], marker='o', linestyle='-', color='blue')
plt.title("Test Score Trend")
plt.xlabel("Index")
plt.ylabel("Test Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Gender Distribution")
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Test Scores'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Test Scores")
plt.xlabel("Test Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
